```python
movies = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = 'title') # merging both of the data frames on the basis of title
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']] # new dataframe keeping only the imp rows
```

```python
import ast # this library has the function that makes string into literals
```

```python
def convert(text):
  L = []
  for i in ast.literal_eval(text):   # converting the dic item into an imp list which will be used
    L.append(i['name'])
    return L
```

```python
from google.colab import data_table
from vega_datasets import data

data_table.enable_dataframe_formatter()

data.movies()
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
 #cleaning the row for only data that is needed
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:                    # taking only top 3 names
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):  # to just fetch the names of director from crew
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    if L is None or not isinstance(L, list):  # Handle NoneType or non-list values
        return []  # Return an empty list
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))  # Remove spaces
    return L1
```

```python
print(movies['genres'].isnull().sum())  # Count missing values
print(movies['genres'].head())  # View the first few entries
```

```python
movies['genres'].dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
# Check the data type of the 'overview' column
print(movies['overview'].dtype)
```

```python
# Python that splits a string into a list of words
movies['overview'] = movies['overview'].apply(lambda x:x.split() if isinstance(x, str) else x)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] #adding all the columns to make one tag formate
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew']) #dropping unimportant column
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
```

```python
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english') # using count vectorization we can creates vector for each movie on the basis of wordS where the repeated no of word would be given a wieght so the words coming in each film will have wieghts acoordingly
```

```python
vector = cv.fit_transform(new['tags']).toarray() # it willl not be there as a defined form so we need to convert it into an array
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector) # this will calculate all vector distance on the basis of cosine values
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0] # to find the index of that given movie name
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) # sorting the similarity to see the most compatible one with the other movies 1 means hieghest compatible
    for i in distances[1:6]: # getting top 5 movies
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
